```python
# program for bank account 
class bank:
    def __init__(self):
                # Constructor to initialize the bank attributes 

        self.name_depositor =input("enter the name of depositor!\n")
        self.account_number = int(input("enter the account number!\n"))      
        self.account_type = input("enter the name for account type!\n")       
        self.balance = float(input("enter the balance!\n"))         
        
    def deposit_function(self, amount):
                # Deposit some amount into the account

        self.balance +=amount 
        
        
    
    def withdraw_function(self, amount):
                # Withdraw some amount from the account

        self.balance -= amount
    
    
    def display_function(self):
#         display function to display name and balance.
        print(f" name of depositor {self.name_depositor}")
        print(f"current balance {self.balance}")
print("details!")        
HBL_bank = bank()

displayBalance = input("Display Balance? (y/n): ")
if displayBalance.lower() == 'y' :
    HBL_bank.display_function()
    
depositBalance = input("Deposit Balance? (y/n): ")
if depositBalance.lower() == 'y' :
    amount = float(input('Enter Deposit Amount: '))
    HBL_bank.deposit_function(amount)
    HBL_bank.display_function()
        
withdrawBalance = input("Withdraw Balance? (y/n): \n")
if withdrawBalance.lower() == 'y' :
    amount = float(input('Enter Withdraw Amount:\n '))
    HBL_bank.withdraw_function(amount)
    HBL_bank.display_function()
```
